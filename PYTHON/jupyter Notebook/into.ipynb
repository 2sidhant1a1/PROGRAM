```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
plt.title("Customized Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
```

```python
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y, color = "red")
axs[0, 1].hist(x + y, color = "blue")
axs[1, 0].pie(x + y)
axs[1, 1].bar(x, y)
plt.show()
plt.savefig("plot.png")
```
